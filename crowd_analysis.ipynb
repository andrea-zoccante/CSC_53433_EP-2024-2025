```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

```python
# Read the data into a DataFrame
df = pd.read_csv("Assets/AnimalResourceData.csv")

df['Timestamp'] = pd.to_datetime(df['Timestamp'])
window_size = 1000
df['Animals_MA'] = df['NAnimals'].rolling(window=window_size).mean()
df['Resources_MA'] = df['NResources'].rolling(window=window_size).mean()
df['NFemales_MA'] = df['NFemales'].rolling(window=window_size).mean()
df['NMales_MA'] = df['NMales'].rolling(window=window_size).mean()
df['MeanLifeSuccess_MA'] = df['MeanLifeSuccess'].rolling(window=window_size).mean()
df['MeanAge_MA'] = df['MeanAge'].rolling(window=window_size).mean()
```

```python
# Plotting
fig, axes = plt.subplots(1, 2, figsize=(12, 5))


# Plot Number of Animals with moving average
axes[0].plot(df['Timestamp'], df['NAnimals'], label='Number of Animals')
axes[0].plot(df['Timestamp'], df['Animals_MA'], color='red', linestyle='--', label=f'Moving Average ({window_size})')
axes[0].set_title('Number of Animals Over Time')
axes[0].set_ylabel('Number of Animals')
axes[0].grid(True)
axes[0].xaxis.set_major_locator(ticker.NullLocator())
axes[0].legend()

# Plot Number of Resources with moving average
axes[1].plot(df['Timestamp'], df['NResources'], color='orange', label='Number of Resources')
axes[1].plot(df['Timestamp'], df['Resources_MA'], color='green', linestyle='--', label=f'Moving Average ({window_size})')
axes[1].set_title('Number of Resources Over Time')
axes[1].set_ylabel('Number of Resources')
axes[1].grid(True)
axes[1].xaxis.set_major_locator(ticker.NullLocator())
axes[1].legend()

# Adjust layout and show
plt.tight_layout()
plt.show()
```
